# Import Numpy

```python
import numpy as np
```

# Creating Matrix: 2-D Array

```python
# Create the following rank 2 array with shape (3, 2) =3 rows and 2 columns

data = np.array([[1,2], [3,4], [5,6]])
print(data)
```

```python
ones=np.ones((3,2))     # 3 rows and 2 columns with all values of matrix as 1 due to np.ones() 
print(ones)
```

```python
zeros=np.zeros((3,2))   # 3 rows and 2 columns with all values/elemtns of matrix as 0 due to np.zeros() 
print(zeros)
```

```python
random=np.random.random((3,2))    # 3 rows and 2 columns with all values/elemtns of matrix +  
print(random)                      # + as an arbitrary random number generation due to np.random.random(())
```

# Matrix Arithmetic Operations

```python
addition = data + ones 
print(addition)
```

```python
multiplication= data*zeros
print (multiplication)
```

# Matrix Indexing and Slicing

```python
index1=data[0,1]    # data[a,b] a=row b= column separated by comma "  ,  "  
print(index1)       #data[0,1]  = 0th row 1st column value
```

```python
slice1=data[0:2] # data[a:n] row and columns seprated by " , "  but here no comma so its simply a row and all colmn by default   
print(slice1)   # data[a:n] = means all matrix elementsfrom ath row to n-1th row and all columns by default
                    # data[0:2] = means all matrix elementsfrom 0th row to 2-1th = 1st row and all columns by default
```

```python
slice2=data[0:2,1]  # data[0:2] = means all matrix elementsfrom 0th row to 2-1th = 1st row i.e. all elements from 0 to 1st row  
print(slice2)       # and all elemtns of 0th columns 
```

# Matrix Aggregation

```python
print(data.max())
print(data.min())
print(data.sum())
```

# Transposing

```python
dataT=data.T
print(dataT)
```

# Reshaping

```python
data_vertical=data.reshape(6,1)
print(data_vertical)
```

```python
data_horizontal=data.reshape(1,6)
print(data_horizontal)
```

```python
data_landscape=data.reshape(2,3)
print(data_landscape)
```
